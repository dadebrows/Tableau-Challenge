```python
import pandas as pd
```

```python
filename="JC-202003-citibike-tripdata.csv"
bikes_df=pd.read_csv(filename)
```

```python
bikes_df.head()
```

```python
bikes_df.describe()
```

```python
start=['start station latitude','start station longitude','start station name']
start_df=bikes_df.groupby(by=start).count().reset_index()
start_df=start_df[['start station latitude','start station longitude','bikeid']]
start_df=start_df.rename(columns={'start station latitude':'latitude',
                                  'start station longitude':'longitude',
                                  'bikeid':'start_count'})
start_df.head()
```

```python
end=['end station latitude','end station longitude','end station name']
end_df=bikes_df.groupby(by=end).count().reset_index()
end_df=end_df[['end station latitude','end station longitude','end station name','bikeid']]
end_df=end_df.rename(columns={'end station latitude':'latitude',
                                'end station longitude':'longitude',
                                'end station name':'station',
                                'bikeid':'end_count'})
```

```python
merge_vars=['latitude','longitude']
station_df=pd.merge(start_df,end_df,how='outer',on=merge_vars)
station_df.head()
```

```python
station_df.to_csv('summary.csv')
```
